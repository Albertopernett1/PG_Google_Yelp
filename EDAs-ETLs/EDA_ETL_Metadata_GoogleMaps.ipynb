## Google Maps metadata EDA

```python
# Encuentra la ubicacion de spark
import findspark
findspark.init()
import pyspark
findspark.find()
```

```python
# Inicia la sesion de spark
from pyspark import SparkContext, SparkConf
from pyspark.sql import SparkSession
conf = pyspark.SparkConf().setAppName('appName').setMaster('local')
sc = pyspark.SparkContext(conf=conf)
spark = SparkSession(sc)
```

```python
spark
```

```python
import databricks.koalas as ks
import os
```

```python
# Se leen todos los archivos de la carpeta
directorio = r'D:\Proyecto Integrador Parquet\Google Maps\metadata-sitios'
dataframes_koalas = []

# Leer archivos parquet en dataframes de spark
for archivo in os.listdir(directorio):
    if archivo.endswith('.parquet'):
        ruta_archivo = os.path.join(directorio, archivo)
        df_spark = spark.read.parquet(ruta_archivo)
        df_koalas = ks.DataFrame(df_spark)  # Convertir a dataframe de koalas
        dataframes_koalas.append(df_koalas.drop("MISC"))  # Elimina la columna MISC

# Concatenar dataframes de koalas
df = ks.concat(dataframes_koalas, ignore_index=True)
```

```python
# Elimina las columnas description y price porque contienen pocos valores
df = df.drop(columns=['description', 'price'])
```

```python
# Conteo de valores nulos
df.isna().sum()
```

```python
# Estadisticas descriptivas de las variables numericas
df.describe()
```

```python
# Conteo de negocios por categoria
# Convertir las listas en filas individuales utilizando explode
df_exploded = df.explode('category')

# Obtener el conteo de valores únicos
value_counts = df_exploded['category'].value_counts()

print(value_counts)
```
